```python
#Numpy Fuction
def add(a1,a2):
  return a1+a2

add_num=np.frompyfunc(add,2,1)
add_num([1,2,3],[4,5,6])
```

```python
def mult(a,b,c):
  return a*b+c

mult_num=np.frompyfunc(mult,3,1)
```

```python
mult_num([1,2,3],[4,5,6],[7,8,9])
```

```python
mult_num(3,2,3)
```

```python
import matplotlib.pyplot as plt
import numpy as np


y=np.arange(0,10)*np.random.randint(0,10,10)

plt.plot(y,marker='v')
plt.show()
```

```python
x=np.array([0,4,5,1,6,7])
y=np.array([0,1,5,1,2,7])

plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title(label="Sample Graph")

plt.plot(x,y,marker='o',color='red',label='line 1', linewidth=2)
plt.plot(y,marker='o',color='blue',label='line 2')
plt.plot()
plt.legend()
plt.grid()

plt.show()
```

```python
x=np.array([0,2,3,7,9,12])
y=np.array([0,4,2,6,5,10])
plt.figure(figsize=(20,20),edgecolor='Red')
plt.plot(x,y,marker='.',label='ABC')
plt.plot(x+1,y+2,marker='v',label='BAC')
plt.xlabel('Time')
plt.ylabel('Prie')
plt.title('Stock Prize')
plt.legend()
plt.grid()
plt.show()
```
