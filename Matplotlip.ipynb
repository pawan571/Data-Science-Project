```python
#sactter
x=np.random.normal(5,5,1000)
y=np.random.normal(10,4,1000)
col=np.random.randint(1,100,1000)
plt.scatter(x,y, c=col, cmap='viridis')
plt.show()
```

```python
x=np.random.normal(5,5,1000)
y=np.random.normal(10,4,1000)
col=np.random.randint(1,100,1000)
col_2=(x.max()-x/(x.max()-x.min()))
plt.scatter(x,y, c=col_2, cmap='viridis')
plt.show()
```
